In this notebook I performed the analysis to answer the second question:

Which country there is more professional?

```python
import numpy as np
import pandas as pd

```

```python
df = pd.read_csv('Desktop/documentos/projetos/data_science/nano/project_one/survey_results_public_2020.csv')

# this the same filter of the last notebook. I reused here because, this analyses focus only professional programmers 

df = df[df['MainBranch'] == 'I am a developer by profession']
```

UNDERSTADING DATA 

```python
df["Country"].value_counts()
```


It is possible to see that there are more professionals in the USA, but this damage shows the professionals as a whole, we will see when we separate by graduation what happens. Let's prepare the data


```python
# here I clean the data  

df['Country'].isnull().sum()
```

```python
df['Country'].dropna()
```

```python
# First I take the total of programmers by country 

prof_1 = df["Country"].value_counts().reset_index()
prof_1
```

```python
#take the professional developer that gratuated in other area, around the world. 

df_by_country =df[df['UndergradMajor'] != 'Computer science, computer engineering, or software engineering']
```

```python
prof_2 = df_by_country["Country"].value_counts().reset_index()

prof_2
```

```python
# I use this, and the next cell,  to make a new DataFrame with the professionals in TI major and Other areas

df_profession = pd.DataFrame(prof_1.rename(columns={'index' : "Country", "Country":'Total'}))

df_profession
```

```python
No_TI= pd.DataFrame(prof_2.rename(columns={'index': "Country", "Country":'Other Area'}))

No_TI
```


MAKING ANALYSIS AND OBTAINING STATISTICS


```python
# I make in this to think easier for put all columns together 

df_profession['Only TI']=  df_profession['Total'] - No_TI['Other Area']

df_profession
```

```python
df_profession["Other"] = No_TI['Other Area']

df_profession
```

```python
# See the procentagem of professionals with other areas working how professional developer.

df_profession['Mean'] = (df_profession["Other"]* 100)/ df_profession['Total']

df_profession
```

```python
df_profession.dropna()
```

```python
# this cell and the next, it is just for curiosity.

df_profession[df_profession['Mean'] == df_profession['Mean'].min()]
```

```python
df_profession[df_profession['Mean'] == df_profession['Mean'].max()]
```


As a result, it is possible to see that parents with a greater number of professional programmers remain, and there is a great balance between professionals with degrees in Computer Science and professionals with degrees in other areas, when we look at these other areas as a single group.
